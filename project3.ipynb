```python
import cv2
import os
from matplotlib import pyplot as plt
```

```python
img = cv2.imread(os.path.join(path,'1.jpg'))
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
median = cv2.medianBlur(gray,3)
```

```python
plt.imshow(median,cmap='gray')
```

```python
plt.hist(median.flat,bins=100,range=(0,255))
```

```python
gauss = cv2.GaussianBlur(median,(3,3),0)
```

```python
plt.imshow(gauss,cmap='gray')
```

```python
plt.hist(gauss.flat,bins=100,range=(0,255))
```

```python
canny = cv2.Canny(gauss,115,150)
```

```python
plt.imshow(canny,cmap='gray')
```

```python
import numpy as np
```

```python
kernel = np.ones((6,6),np.uint8)
```

```python
close = cv2.morphologyEx(canny,cv2.MORPH_CLOSE,kernel)
```

```python
plt.imshow(close,cmap='gray')
```

```python
dilation = cv2.dilate(close,kernel,iterations=3)
```

```python
plt.imshow(dilation,cmap='gray')
```

```python
erosion = cv2.erode(dilation,kernel,iterations=3)
```

```python
plt.imshow(erosion,cmap='gray')
```

```python
dilation2 = cv2.dilate(erosion,kernel,iterations=2)
```

```python
plt.imshow(dilation2,cmap='gray')
```

```python
close2 = cv2.morphologyEx(dilation2,cv2.MORPH_CLOSE,kernel)
```

```python
plt.imshow(close2,cmap='gray')
```

```python
erosion2 = cv2.erode(close2,kernel,iterations=2)
```

```python
plt.imshow(erosion2,cmap='gray')
```
