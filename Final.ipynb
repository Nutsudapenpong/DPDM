1. Data Mining คือ การทำเหมืองข้อมูล สกัดเอาความรู้ที่ซ่อนอยู่ในข้อมูลออกมา
2. Relational records เป็นการใช้ Index หรือ ID ใช้เชื่อมข้อมูลของตารางนึงไปยังตารางนึง เช่น เลขบัตรประชาชน
3. Data matrix เป็นการเก็บข้อมูลที่ตารางมีหัวข้อสองฝั่ง
4. Transaction data เป็นการเก็บข้อมูลที่ไม่มีหัวข้อ แค่เรียงๆ ขั้นด้วย comma
5.Document data เป็นการเก็บข้อมูลที่มี attributes ด้านบน และ ตัวบอกตัวตนด้านข้าง เหมือน ตารางทั่วไป
6. symmetric data เป็นข้อมูลที่มีความสำคัญเท่ากันทั้ง 2 ข้าง 
7. Asymmetric เป็นข้อมูลที่มีค่าความสำคัญสองฝั่งไม่เท่ากัน เช่น ทางการแพทย์ จะสนใจฝั่งที่เป็นโรคมากกว่า

9. Data Reduction and Transformation เป็นวิธีการลดข้อมูลทั้งฝั่ง feature และ Data
10. Mining frequent pattern = การหารูปแบบที่เกิดขึ้นบ่อย ๆ ซ้ำ ๆ
11. absolute support = เป็นการนับจำนวนเป๊ะ ๆ ของ support แต่จะไม่รู้ว่าทั้งหมดมีการซื้อกี่ครั้ง เลยดูไม่ได้ว่าจำนวนเท่านี้คือมากหรือน้อย
12. relative support = เป็นการแสดงผลลัพธ์บอกแบบเป็นเปอร์เซ็นของ support รู้ว่ามีการซื้อทั้งกี่ครั้ง
13. minsup = item ที่ผ่านค่า support ขั้นต่ำ
14. threshold = ตัวที่ใช้แบ่งว่าอันนี้ผ่านเกณฑ์ ไม่ผ่านเกณฑ์ เช่น support เกิน 50% เอา
15. Decision tree เป็นการ classification ที่ใช้การคำนวณเพื่อใช้พิจารณา node และใช้ node ในการแบ่งข้อมูลให้อยู่ในกลุ่มที่ถูกต้อง
16. Inter-quartile range คือ ความยาวของแท่ง Boxplot เกิดจาก Q3-Q1
17. whisker คือ เส้น 1.5 เท่าของ Inter-quatile range ถ้าหลุดคืออยู่นอก outliers
18. Outliers คือ distribution ผิดพลาด อาจเกิดจากการกรอกข้อมูลที่ผิดพลาด
