```
y = New_world2[['Life expectancy']]
y
```

```
from sklearn.neighbors import KNeighborsClassifier
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2) 
```

```
print(f'{x_train.shape} {x_test.shape}') 
```

```
from sklearn.model_selection import cross_val_score 
```

```
KNN_k3_wU = KNeighborsClassifier(n_neighbors=3) #Model1
```

```
cross_val_score(KNN_k3_wU, x_train, y_train, cv=5) 
cross_val_score(KNN_k3_wU, x_train, y_train, cv=5).mean() 
```

```
KNN_k3_wD = KNeighborsClassifier(n_neighbors=3, weights='distance') #Model2
cross_val_score(KNN_k3_wD, x_train, y_train, cv=5).mean()
```

```
KNN_k3_wF = KNeighborsClassifier(n_neighbors=3, weights='uniform') #Model3
cross_val_score(KNN_k3_wF, x_train, y_train, cv=5).mean()
```

```
KNN_k3_wH = KNeighborsClassifier(n_neighbors=4, weights='distance') #Model4
cross_val_score(KNN_k3_wH, x_train, y_train, cv=5).mean()
```

```
KNN_k3_wG = KNeighborsClassifier(n_neighbors=4, weights='uniform') #Model5
cross_val_score(KNN_k3_wG, x_train, y_train, cv=5).mean()
```

```
KNN_k3_wI = KNeighborsClassifier(n_neighbors=5, weights='distance') #Model6
cross_val_score(KNN_k3_wI, x_train, y_train, cv=5).mean()
```

```
KNN_k3_wI = KNeighborsClassifier(n_neighbors=5, weights='distance') #The best model
KNN_k3_wI.fit(x_train,y_train)
KNN_k3_wI.score(x_test,y_test)
```
